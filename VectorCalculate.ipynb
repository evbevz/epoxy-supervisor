```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm
```

```python
#@title Наглядное отображение операций с векторами

# Создаем вектора
vA0 = np.array([0,0])
vA1 = np.array([6,2])
vA2 = np.array([8,8])
vX  = np.array([6,3])

vA1X  = vX - vA1
vA2X  = vX - vA2
vA1A2 = vA2 - vA1 

# Вычисляем L2-норму (Евклидову метрику, Евклидово расстояние)
L2_A1   = norm(vA1,ord=2)
L2_A2   = norm(vA2,ord=2)
L2_X    = norm(vX,ord=2)
L2_A1A2 = norm(vA1A2,ord=2)
L2_A1X  = norm(vA1X,ord=2)
L2_A2X  = norm(vA2X,ord=2)

# Вывод
print ("vA1:", vA1, "\tL2:", L2_A1)
print ("vA2:", vA2, "\tL2:", L2_A2)
print ("vX:", vX, "\tL2:", L2_X)
print ("vA1A2:", vA1A2, "\tL2:", L2_A1A2)
print ("vA1X:", vA1X, "\tL2:", L2_A1X)
print ("vA2X:", vA2X, "\tL2:", L2_A2X)

# Высчитываем угол между векторами
#cos_angle = np.dot(vX-vA1, vA2-vA1) / norm(vX-vA1) / norm(vA2-vA1)
#print ('Косинус угла между A1X и A1A2:', cos_angle)
#print ('Сам угол:', np.degrees(np.arccos(cos_angle)))
pr_A1XonA1A2 = np.dot(vA1X, vA1A2) / norm(vA1A2)
print ('Проекция A1X на A1A2:', pr_A1XonA1A2)
pr_A2XonA1A2 = np.dot(vA2X, vA1A2) / norm(vA1A2)
print ('Проекция A2X на A1A2:', pr_A2XonA1A2)
pr_XonA1 = np.dot(vX, vA1) / norm(vA1)
print ('Проекция X на A1:', pr_XonA1)
pr_A1XonA1 = np.dot(vA1X, vA1) / norm(vA1)
print ('Проекция A1X на A1:', pr_A1XonA1)

# Визуализация
arrowprops = {'arrowstyle': '<|-'}
plt.annotate(r'$\vec A_1$',xy=(0,0),xytext=vA1,arrowprops=arrowprops|{'color': '#000000'})
plt.annotate(r'$\vec A_2$',xy=(0,0),xytext=vA2,arrowprops=arrowprops|{'color': '#00ff00'})
plt.annotate('',xy=vA1,xytext=vA2,arrowprops=arrowprops|{'color': '#00aa00'})
plt.annotate(r'$\vec X$',xy=(0,0),xytext=vX,arrowprops=arrowprops|{'color': '#ff0000'})
plt.annotate('',xy=vA1,xytext=vX,arrowprops=arrowprops|{'color': '#aa0000'})
plt.annotate('',xy=vA2,xytext=vX,arrowprops=arrowprops|{'color': '#aa00aa'})

plt.axis('square')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.grid(linewidth=0.5)
plt.show()
```
