```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import norm
```

```python
#@title Пример визуализации векторов

# Создаем вектора
vA0 = np.array([0,0])
vA1 = np.array([2,1])
vA2 = np.array([4,3])
vA3 = np.array([5,5])

# Визуализация
arrowprops = {'arrowstyle': '<|-'}
plt.arrow(0,0,2,1)
plt.annotate(r'$\vec A2$',xy=(0,0),xytext=vA2,arrowprops={'arrowstyle': '<|-'})
plt.quiver(*vA3,scale=10)
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid(True)
plt.show()

```

```python
#@title Наглядное отображение операций с векторами

# Создаем вектора
vA0 = np.array([0,0])
vA1 = np.array([5,3])
vA2 = np.array([7,5])
vA3 = np.array([6,8])
vX = np.array([4,2])

# Вычисляем L1-норму (Манхэттенское расстояние)
L1_A1 = norm(vA1,ord=1)
L1_A2 = norm(vA2,ord=1)
L1_A3 = norm(vA3,ord=1)
L1_X = norm(vX,ord=1)
L1_A1A2 = norm(vA2-vA1,ord=1)
L1_A2A3 = norm(vA3-vA2,ord=1)
L1_A1X = norm(vX-vA1,ord=1)
L1_XA2 = norm(vA2-vX,ord=1)

# Вычисляем L2-норму (Евклидову метрику, Евклидово расстояние)
L2_A1 = norm(vA1,ord=2)
L2_A2 = norm(vA2,ord=2)
L2_A3 = norm(vA3,ord=2)
L2_X = norm(vX,ord=2)
L2_A1A2 = norm(vA2-vA1,ord=2)
L2_A2A3 = norm(vA3-vA2,ord=2)
L2_A1X = norm(vX-vA1,ord=2)
L2_XA2 = norm(vA2-vX,ord=2)

# Вывод
print ("vA1:", vA1, "\tL1:", L1_A1, "\tL2:", L2_A1)
print ("vA2:", vA2, "\tL1:", L1_A2, "\tL2:", L2_A2)
print ("vA3:", vA3, "\tL1:", L1_A3, "\tL2:", L2_A3)
print ("vX:", vX, "\tL1:", L1_X, "\tL2:", L2_X)
print ("vA1A2:", vA2-vA1, "\tL1:", L1_A1A2, "\tL2:", L2_A1A2)
print ("vA2A3:", vA3-vA2, "\tL1:", L1_A2A3, "\tL2:", L2_A2A3)
print ("vA1X:", vX-vA1, "\tL1:", L1_A1X, "\tL2:", L2_A1X)
print ("vXA2:", vA2-vX, "\tL1:", L1_XA2, "\tL2:", L2_XA2)

# Высчитываем угол между векторами
#cos_angle = np.dot(vX-vA1, vA2-vA1) / norm(vX-vA1) / norm(vA2-vA1)
#print ('Косинус угла между A1X и A1A2:', cos_angle)
#print ('Сам угол:', np.degrees(np.arccos(cos_angle)))
pr_XonA1A2 = np.dot(vX-vA1, vA2-vA1) / norm(vA2-vA1)
print ('Проекция X на A1A2:', pr_XonA1A2)

# Визуализация
arrowprops = {'arrowstyle': '<|-'}
plt.annotate(r'$\vec A_1$',xy=(0,0),xytext=vA1,arrowprops=arrowprops|{'color': '#000000'})
plt.annotate(r'$\vec A_2$',xy=(0,0),xytext=vA2,arrowprops=arrowprops|{'color': '#00ff00'})
plt.annotate('',xy=vA1,xytext=vA2,arrowprops=arrowprops|{'color': '#00aa00'})
plt.annotate(r'$\vec A_3$',xy=(0,0),xytext=vA3,arrowprops=arrowprops|{'color': '#0000ff'})
plt.annotate('',xy=vA2,xytext=vA3,arrowprops=arrowprops|{'color': '#0000aa'})
plt.annotate(r'$\vec X$',xy=(0,0),xytext=vX,arrowprops=arrowprops|{'color': '#ff0000'})
plt.annotate('',xy=vA1,xytext=vX,arrowprops=arrowprops|{'color': '#aa0000'})

plt.axis('square')
plt.xlim(0, 8)
plt.ylim(0, 8)
plt.grid(linewidth=0.5)
plt.show()

```
