```python
from lxml import etree
import matplotlib.pyplot as plt
import os.path
import torch
from ultralytics import YOLO

# ÐšÐ¾ÑÐ¸Ð½ÑƒÑÐ½Ð¾Ðµ ÑÑ…Ð¾Ð´ÑÑ‚Ð²Ð¾ (ÐšÐ¾ÑÐ¸Ð½ÑƒÑ ÑƒÐ³Ð»Ð° Ð¼ÐµÐ¶Ð´Ñƒ Ð²ÐµÐºÑ‚Ð¾Ñ€Ð°Ð¼Ð¸)
def cos_sim (v1, v2):
    cosine_similarity = torch.dot(v1, v2) / (torch.linalg.vector_norm(v1) * torch.linalg.vector_norm(v2))
    return cosine_similarity

# ÐŸÑ€Ð¾ÐµÐºÑ†Ð¸Ñ Ð²ÐµÐºÑ‚Ð¾Ñ€Ð° Ð½Ð° Ð²ÐµÐºÑ‚Ð¾Ñ€
def pr_vec (v1, v2):
    pr_v1_on_v2 = torch.dot(v1, v2) / torch.linalg.vector_norm(v2)
    return pr_v1_on_v2

# Ð—Ð°Ð³Ñ€ÑƒÐ·ÐºÐ° Ñ„Ð°Ð¹Ð»Ð° Ð°Ð½Ð¾Ñ‚Ð°Ñ†Ð¸Ð¸ Ð² Ñ„Ð¾Ñ€Ð¼Ð°Ñ‚Ðµ "CVAT for images 1.1"
# Ð˜Ð· Ñ„Ð°Ð¹Ð»Ð° Ð±ÐµÑ€Ñ‘Ñ‚ÑÑ Ð¼Ð°ÑÑÐ¸Ð² Ñ‚Ð¾Ñ‡ÐµÐº Ð¿Ð¾Ð»Ð¸Ð»Ð¸Ð½Ð¸Ð¸ Ñ Ñ€Ð°Ð·Ð¼ÐµÑ‚ÐºÐ¾Ð¹ ÑˆÐºÐ°Ð»Ñ‹ Ð´ÐµÐ»ÐµÐ½Ð¸Ñ: 0Ð¼Ð», 1Ð¼Ð», ... 20Ð¼Ð»
def Calibrate (fAnnotation):
    tree = etree.parse(fAnnotation)
    root = tree.getroot()
    result = root.xpath('//polyline[@label="injector meter"]')
    if not(result):
        raise Exception("ÐÐµ ÑƒÐ´Ð°Ð»Ð¾ÑÑŒ Ð½Ð°Ð¹Ñ‚Ð¸ polyline Ñ Ñ‚Ð¾Ñ‡ÐºÐ°Ð¼Ð¸ ÑƒÑ€Ð¾Ð²Ð½ÐµÐ¹ Ð¸Ð½Ð¶ÐµÐºÑ‚Ð¾Ñ€Ð° Ð² Ñ„Ð°Ð¹Ð»Ðµ: ", fAnnotation)
    points = result[0].attrib['points']
    # Ð¡Ð»Ð¾Ð¶Ð½Ð°Ñ ÐºÐ¾Ð½ÑÑ‚Ñ€ÑƒÐºÑ†Ð¸Ñ, Ð½Ð¾ Ð¿Ð¾ ÑÑƒÑ‚Ð¸ Ñ€Ð°Ð·Ð±Ð¸Ñ€Ð°ÐµÑ‚ ÑÑ‚Ð¾Ñ€Ð¾ÐºÑƒ Ð½Ð° ÑÐ¿Ð¸ÑÐ¾Ðº ÑÐ¿Ð¸ÑÐºÐ¾Ð² (Ð´Ð²ÑƒÐ¼ÐµÑ€Ð½Ñ‹Ð¹ ÑÐ¿Ð¸ÑÐ¾Ðº) Ð¸Ð· Ð´Ð²ÑƒÑ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð¿Ð¾Ð¿ÑƒÑ‚Ð½Ð¾ Ð¿ÐµÑ€ÐµÐ²Ð¾Ð´Ñ Ð¸Ð· ÑÑ‚Ñ€Ð¾ÐºÐ¾Ð²Ñ‹Ñ… Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð² Ñ‡Ð¸ÑÐ»Ð¾Ð²Ð¾Ðµ Ð·Ð½Ð°Ñ‡Ð½Ð¸Ðµ.
    list_of_points = list(map(lambda x: list(map(float,x.split(','))), points.split(';')))
    array_of_points = torch.tensor(list_of_points)
    #print ("ÐœÐ°ÑÑÐ¸Ð² Ñ‚Ð¾Ñ‡ÐµÐº ÑˆÐºÐ°Ð»Ñ‹ Ð¸Ð½Ð¶ÐµÐºÑ‚Ð¾Ñ€Ð°:\n", array_of_points)
    return array_of_points

# ÐŸÐ¾Ð»ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ñ†ÐµÐ½Ñ‚Ñ€Ð° ÑÐ»Ð»Ð¸Ð¿ÑÐ° Ð¿Ð¾ Ñ‚Ð¾Ñ‡ÐºÐ°Ð¼: Ð»ÐµÐ²Ð¾Ð¹ Ð¸ Ð¿Ñ€Ð°Ð²Ð¾Ð¹ Ð±Ð¾Ð»ÑŒÑˆÐ¾Ð¹ Ð¾ÑÐ¸, Ð´Ð°Ð»ÑŒÐ½ÐµÐ¹ Ð¸ Ð±Ð»Ð¸Ð¶Ð½ÐµÐ¹ Ð¼Ð°Ð»Ð¾Ð¹ Ð¾ÑÐ¸.
# Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ Ñ‡ÐµÑ€ÐµÐ· ÑÐµÑ€ÐµÐ´Ð¸Ð½Ñƒ Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»ÐµÐ¹ ÑÐ»Ð»Ð¸Ð¿ÑÐ°. ÐÐ° Ð±ÑƒÐ´ÑƒÑ‰ÐµÐµ ÑÑ‚Ð¾Ð¸Ñ‚ Ð¿Ñ€Ð¾ÑÑ‚Ð¾ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÑÑ‚ÑŒ bounding box ÑÑ‚Ð¾Ð³Ð¾ ÑÐ»Ð»Ð¸Ð¿ÑÐ°. ÐœÐ¾Ð´ÐµÐ»ÑŒ Ð±ÑƒÐ´ÐµÑ‚ Ð²Ñ‹Ð´Ð°Ð²Ð°Ñ‚ÑŒ ÑƒÐ¶Ðµ ÐºÐ°Ðº Ñ€Ð°Ð· ÑÐµÑ€ÐµÐ´Ð¸Ð½Ñƒ.
def GetEllipseCenter (kpt, kptConfidence):
    #print ("Ð›ÐµÐ²Ð°Ñ Ñ‚Ð¾Ñ‡ÐºÐ° Ð±Ð¾Ð»ÑŒÑˆÐ¾Ð¹ Ð¾ÑÐ¸:\t", kpt[0])
    #print ("ÐŸÑ€Ð°Ð²Ð°Ñ Ñ‚Ð¾Ñ‡ÐºÐ° Ð±Ð¾Ð»ÑŒÑˆÐ¾Ð¹ Ð¾ÑÐ¸:\t", kpt[1])
    #print ("Ð”Ð°Ð»ÑŒÐ½ÑÑ Ñ‚Ð¾Ñ‡ÐºÐ° Ð¼Ð°Ð»Ð¾Ð¹ Ð¾ÑÐ¸:\t", kpt[2])
    #print ("Ð‘Ð»Ð¸Ð¶Ð½ÑÑ Ñ‚Ð¾Ñ‡ÐºÐ° Ð¼Ð°Ð»Ð¾Ð¹ Ð¾ÑÐ¸:\t", kpt[3])

    if len(kpt) == 0:
        return None

    # ÐŸÑ€Ð¾Ð²ÐµÑ€ÑÐµÐ¼ ÑƒÐ²ÐµÑ€ÐµÐ½Ð½Ð¾ÑÑ‚ÑŒ Ð² Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ð¸ Ñ‚Ð¾Ñ‡ÐºÐ¸. Ð•ÑÐ»Ð¸ Ð¿Ð¾ÐºÐ°Ð·Ð°Ñ‚ÐµÐ»ÑŒ Ð¼ÐµÐ½ÑŒÑˆÐµ Ð·Ð°Ð´Ð°Ð½Ð½Ð¾Ð³Ð¾ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ, Ñ‚Ð¾ Ñ‚Ð¾Ñ‡ÐºÑƒ Ð¸Ð³Ð½Ð¾Ñ€Ð¸Ñ€ÑƒÐµÐ¼.
    # Ð¡ÐµÑ€ÐµÐ´Ð¸Ð½Ð° Ð¼ÐµÐ¶Ð´Ñƒ Ð»ÐµÐ²Ð¾Ð¹ Ð¸ Ð¿Ñ€Ð°Ð²Ð¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐ°Ð¼Ð¸ Ð±Ð¾Ð»ÑŒÑˆÐ¾Ð¹ Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»Ð¸ ÑÐ»Ð»Ð¸Ð¿ÑÐ°
    if (kpt[0][2] > kptConfidence) and (kpt[1][2] > kptConfidence):
        big_axis_center = (kpt[0][0:2:] + kpt[1][0:2:]) / 2
    else:
        big_axis_center = None
    # Ð¡ÐµÑ€ÐµÐ´Ð¸Ð½Ð° Ð¼ÐµÐ¶Ð´Ñƒ Ð´Ð°Ð»ÑŒÐ½ÐµÐ¹ Ð¸ Ð±Ð»Ð¸Ð¶Ð½ÐµÐ¹ Ñ‚Ð¾Ñ‡ÐºÐ°Ð¼Ð¸ Ð¼Ð°Ð»Ð¾Ð¹ Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»Ð¸ ÑÐ»Ð»Ð¸Ð¿ÑÐ°
    if (kpt[2][2] > kptConfidence) and (kpt[3][2] > kptConfidence):
        small_axis_center = (kpt[2][0:2:] + kpt[3][0:2:]) / 2
    else:
        small_axis_center = None
    # Ð¡ÐµÑ€ÐµÐ´Ð¸Ð½Ð° Ð¼ÐµÐ¶Ð´Ñƒ ÑÐµÑ€ÐµÐ´Ð¸Ð½Ð¾Ð¹ Ð±Ð¾Ð»ÑŒÑˆÐ¾Ð¹ Ð¸ ÑÐµÑ€ÐµÐ´Ð¸Ð½Ð¾Ð¹ Ð¼Ð°Ð»Ð¾Ð¹ Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»ÐµÐ¹. Ð˜Ð½Ð¾Ð³Ð´Ð° Ð¼Ð¾Ð¶ÐµÑ‚ Ð±Ñ‹Ñ‚ÑŒ Ð½Ðµ Ñ€Ð°Ð²Ð½Ð°Ñ ÑÐµÑ€ÐµÐ´Ð¸Ð½Ð°Ð¼ ÐºÐ°Ð¶Ð´Ð¾Ð¹ Ð¸Ð· Ð½Ð¸Ñ… (Ð¿Ñ€Ð¸ Ð¾ÑˆÐ¸Ð±ÐºÐ°Ñ… Ð² Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ð¸ ÐºÐ»ÑŽÑ‡ÐµÐ²Ñ‹Ñ… Ñ‚Ð¾Ñ‡ÐµÐº Ð¼Ð¾Ð´ÐµÐ»ÑŒÑŽ).
    if (big_axis_center is not None) and (small_axis_center is not None):
        ellipse_center = (big_axis_center + small_axis_center) / 2
    elif (big_axis_center is not None) and (small_axis_center is None):
        ellipse_center = big_axis_center
    elif (big_axis_center is None) and (small_axis_center is not None):
        ellipse_center = small_axis_center
    else:
        ellipse_center = None
    #print ("BAC: ", big_axis_center, "SAC: ", small_axis_center, "ELC:", ellipse_center)
    return ellipse_center

# ÐŸÐ¾Ð»ÑƒÑ‡ÐµÐ½Ð¸Ðµ ÑƒÑ€Ð¾Ð²Ð½Ñ ÑÐ¿Ð¾ÐºÑÐ¸Ð´ÐºÐ¸ Ð¿Ð¾ ÐºÐ°Ñ€Ñ‚Ð¸Ð½ÐºÐµ
def GetEpoxyLevel (model, arrayEpoxyLevel, filenameInjectorCam, kptConfidence):
    # Ð—Ð°Ð¿ÑƒÑÐºÐ°ÐµÐ¼ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ðµ
    #results = model.predict(source=filenameInjectorCam, verbose=True)  # ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ðµ Ð¿Ð¾ Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸ÑŽ. Ð’Ð¾Ð·Ð²Ñ€Ð°Ñ‰Ð°ÐµÑ‚ÑÑ ÑÐ¿Ð¸ÑÐ¾Ðº Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð² (Ñ‚.Ðº. Ð¼Ð¾Ð¶Ð½Ð¾ Ð¿ÐµÑ€ÐµÐ´Ð°Ñ‚ÑŒ ÑÐ¿Ð¸ÑÐ¾Ðº ÐºÐ°Ð´Ñ€Ð¾Ð² Ð¸Ð»Ð¸ Ð´Ð°Ð¶Ðµ Ð²Ð¸Ð´ÐµÐ¾)
    results = model.track(source=filenameInjectorCam, verbose=True, show=True)  # ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ðµ Ð¿Ð¾ Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸ÑŽ. Ð’Ð¾Ð·Ð²Ñ€Ð°Ñ‰Ð°ÐµÑ‚ÑÑ ÑÐ¿Ð¸ÑÐ¾Ðº Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð² (Ñ‚.Ðº. Ð¼Ð¾Ð¶Ð½Ð¾ Ð¿ÐµÑ€ÐµÐ´Ð°Ñ‚ÑŒ ÑÐ¿Ð¸ÑÐ¾Ðº ÐºÐ°Ð´Ñ€Ð¾Ð² Ð¸Ð»Ð¸ Ð´Ð°Ð¶Ðµ Ð²Ð¸Ð´ÐµÐ¾)
    print ('--- Results: ---\n', results)
    for i, result in enumerate(results):
        print ('--- Boxes: ---\n', i, ': ', result.boxes)
        print ('--- Keypoints: ---\n', i, ': ', result.keypoints)

    #print ('--- Box0: ---\n', result[0].boxes)
    #print ('--- Box1: ---\n', result[1].boxes)
    # Ð¢ÐµÐ¾Ñ€ÐµÑ‚Ð¸Ñ‡ÐµÑÐºÐ¸ Ð¼Ð¾Ð¶ÐµÑ‚ Ð±Ñ‹Ñ‚ÑŒ ÑÐ¿Ð¸ÑÐ¾Ðº Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð², Ð½Ð¾ Ð±ÐµÑ€Ñ‘Ð¼ Ñ‚Ð¾Ð»ÑŒÐºÐ¾ Ð¾Ð´Ð½Ð¾ - Ð¿ÐµÑ€Ð²Ð¾Ðµ.
    keypoints = results[0].keypoints  # Keypoints object for pose outputs
    #results[0].show()  # display to screen
    #print ('--- Keypoints: ---\n', keypoints)

    # Ð’Ñ‹ÑÑ‡Ð¸Ñ‚Ñ‹Ð²Ð°ÐµÐ¼ Ñ†ÐµÐ½Ñ‚Ñ€ ÑÐ»Ð»Ð¸Ð¿ÑÐ°
    # ÐŸÐµÑ€ÐµÐ´Ð°ÐµÐ¼ Ð² Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ðµ 4 Ñ‚Ð¾Ñ‡ÐºÐ¸ Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»ÐµÐ¹ ÑÐ»Ð»Ð¸Ð¿ÑÐ° Ð² Ð²Ð¸Ð´Ðµ Ñ‚ÐµÐ½Ð·Ð¾Ñ€Ð° 4x3 [[x1,y2,confidence1],[]...] Ð¸ ÐºÐ¾ÑÑ„Ñ„Ð¸Ñ†Ð¸ÐµÐ½Ñ‚ ÑƒÐ²ÐµÑ€ÐµÐ½Ð½Ð¾ÑÑ‚Ð¸ Ð² Ð¿Ñ€Ð°Ð²Ð¸Ð»ÑŒÐ½Ð¾ÑÑ‚Ð¸ Ñ€Ð°ÑÐ¿Ð¾Ð·Ð½Ð°Ð²Ð°Ð½Ð¸Ñ, Ð½Ð¸Ð¶Ðµ ÐºÐ¾Ñ‚Ð¾Ñ€Ð¾Ð³Ð¾ Ð½Ðµ Ð±ÑƒÐ´ÐµÐ¼ ÑÑ‡Ð¸Ñ‚Ð°Ñ‚ÑŒ, Ñ‡Ñ‚Ð¾ Ñ‚Ð¾Ñ‡ÐºÐ¸ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»Ð¸Ð»Ð¸ÑÑŒ Ð¿Ñ€Ð°Ð²Ð¸Ð»ÑŒÐ½Ð¾. Ð¢.Ðµ. ÐºÐ¾Ð¾Ñ€Ð´Ð¸Ð½Ð°Ñ‚Ñ‹ Ñ‚Ð°ÐºÐ¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐ¸ Ð±ÑƒÐ´ÐµÐ¼ ÑÑ‡Ð¸Ñ‚Ð°Ñ‚ÑŒ Ð»Ð¾Ð¶Ð½Ñ‹Ð¼Ð¸ Ð¸ Ñ‚Ð¾Ñ‡ÐºÑƒ Ð¸Ð³Ð½Ð¾Ñ€Ð¸Ñ€Ð¾Ð²Ð°Ñ‚ÑŒ.
    ellipse_center = GetEllipseCenter(keypoints.data[0][0:4:], kptConfidence)

    if ellipse_center is not None:
        # ÐŸÐµÑ€ÐµÐ½Ð¾ÑÐ¸Ð¼ Ð¼Ð°ÑÑÐ¸Ð² Ñ‚Ð¾Ñ‡ÐµÐº ÑˆÐ¿Ñ€Ð¸Ñ†Ð° Ð½Ð° Ñ‚Ð¾ Ð¶Ðµ ÑƒÑÑ‚Ñ€Ð¾Ð¹ÑÑ‚Ð²Ð¾ Ñ€Ð°ÑÑÑ‡ÐµÑ‚Ð° Ð³Ð´Ðµ Ð¸ Ñ‚ÐµÐ½Ð·Ð¾Ñ€Ñ‹ Ð¼Ð¾Ð´ÐµÐ»Ð¸ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ð¹. Ð•ÑÐ»Ð¸ Ñ€Ð°ÑÑ‡Ñ‘Ñ‚Ñ‹ Ð²ÐµÐ»Ð¸ÑÑŒ Ð½Ð° CUDA, Ñ‚Ð¾ Ð»ÑƒÑ‡ÑˆÐµ Ñ‚Ð°Ð¼ Ð¸ ÑÑ‡Ð¸Ñ‚Ð°Ñ‚ÑŒ Ð²ÑÑ‘ Ð¾ÑÑ‚Ð°Ð»ÑŒÐ½Ð¾Ðµ.
        arrayEpoxyLevel = arrayEpoxyLevel.to(ellipse_center.device)

        # Ð’Ñ‹Ñ‡Ð¸ÑÐ»ÑÐµÐ¼ Ð±Ð»Ð¸Ð¶Ð°Ð¹ÑˆÑƒÑŽ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¾Ð²Ð°Ð½Ð½ÑƒÑŽ Ñ‚Ð¾Ñ‡ÐºÑƒ Ðº Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ð¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐµ (ÑÐµÑ€ÐµÐ´Ð¸Ð½Ðµ ÑÐ»Ð»Ð¸Ð¿ÑÐ°)
        LengthMin = (keypoints.orig_shape[0] ** 2 + keypoints.orig_shape[1] ** 2) ** 0.5 # ÐÑƒÐ¶Ð½Ð¾ Ð¿Ñ€Ð¾ÑÑ‚Ð¾ Ð±Ð¾Ð»ÑŒÑˆÐ¾Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ, Ð½Ð¾ Ñ€ÐµÑˆÐ¸Ð» ÑƒÐºÐ°Ð·Ð°Ñ‚ÑŒ Ð¼Ð°ÐºÑÐ¸Ð¼Ð°Ð»ÑŒÐ½Ð¾ Ð²Ð¾Ð·Ð¼Ð¾Ð¶Ð½Ð¾Ðµ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ Ð½Ð° Ð¸Ð·Ð¾Ð±Ñ€Ð°Ð¶ÐµÐ½Ð¸Ð¸ (Ð´Ð¸Ð°Ð³Ð¾Ð½Ð°Ð»ÑŒ)
        LevelMin = 0
        for level, kpt in enumerate(arrayEpoxyLevel):
            Length = torch.norm(ellipse_center - kpt) # Ð Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ Ð¼ÐµÐ¶Ð´Ñƒ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ð¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐ¾Ð¹ Ð¸ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¾Ð²Ð°Ð½Ð½Ð¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐ¾Ð¹ ÑƒÑ€Ð¾Ð²Ð½Ñ
            if Length < LengthMin:                    # Ð•ÑÐ»Ð¸ Ñ€ÐµÑˆÐ¸Ð¼, Ñ‡Ñ‚Ð¾ Ñ…Ð¾Ñ‚Ð¸ Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ð¿Ñ€Ð¸ Ð¾Ð´Ð¸Ð½Ð°ÐºÐ¾Ð²Ð¾Ð¼ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ð¸ Ð¿Ð¾ÐºÐ°Ð·Ñ‹Ð²Ð°Ð» Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð±Ð¾Ð»ÑŒÑˆÐµÐ³Ð¾ ÑƒÑ€Ð¾Ð²Ð½Ñ, Ñ‚Ð¾ Ñ‚Ð¾Ð³Ð´Ð° Ð¿Ð¾ÑÑ‚Ð°Ð²Ð¸Ñ‚ÑŒ Ð·Ð½Ð°Ðº ÑÑ€Ð°Ð²Ð½ÐµÐ½Ð¸Ñ <=
                LengthMin = Length                    # Ð—Ð°Ð¿Ð¾Ð¼Ð¸Ð½Ð°ÐµÐ¼ Ð¼Ð¸Ð½Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ðµ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ
                LevelMin = level                      # Ð—Ð°Ð¿Ð¾Ð¼Ð¸Ð½Ð°ÐµÐ¼ ÑƒÑ€Ð¾Ð²ÐµÐ½ÑŒ Ðº Ñ‚Ð¾Ñ‡ÐºÐµ ÐºÐ¾Ñ‚Ð¾Ñ€Ð¾Ð³Ð¾ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ðµ Ð¼Ð¸Ð½Ð¸Ð¼Ð°Ð»ÑŒÐ½Ð¾Ðµ
            #print ("Lvl: ", level, "\tCalib pt: ", kpt, "\tPredict pt: ", ellipse_center, "\tLength: ", Length)
        #print ("ÐŸÑ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ñ‹Ð¹ ÑƒÑ€Ð¾Ð²ÐµÐ½ÑŒ ÑÐ¿Ð¾ÐºÑÐ¸Ð´ÐºÐ¸: ", LevelMin)

        # Ð”Ð¾Ñ€Ð°ÑÑ‡Ñ‘Ñ‚ Ð´Ñ€Ð¾Ð±Ð½Ð¾Ð¹ Ñ‡Ð°ÑÑ‚Ð¸ ÑƒÑ€Ð¾Ð²Ð½Ñ.
        # ÐžÐ¿Ñ€ÐµÐ´ÐµÐ»ÑÐµÐ¼ Ñ ÐºÐ°ÐºÐ¾Ð¹ ÑÑ‚Ð¾Ñ€Ð¾Ð½Ñ‹ Ð¾Ñ‚ Ð±Ð»Ð¸Ð¶Ð°Ð¹ÑˆÐµÐ¹ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¾Ð²Ð°Ð½Ð½Ð¾Ð¹ Ñ‚Ð¾Ñ‡ÐºÐ¸ ÑƒÑ€Ð¾Ð²Ð½Ñ Ð½Ð°Ñ…Ð¾Ð´Ð¸Ñ‚ÑÑ Ñ‚Ð¾Ñ‡ÐºÐ° Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ñ Ð¸ Ð½Ð° ÐºÐ°ÐºÐ¾Ð¼ Ñ€Ð°ÑÑÑ‚Ð¾ÑÐ½Ð¸Ð¸
        vSrc = ellipse_center - arrayEpoxyLevel[LevelMin]
        if (LevelMin == (arrayEpoxyLevel.size(dim=0) - 1)): # ÐžÐ³Ñ€Ð°Ð½Ð¸Ñ‡ÐµÐ½Ð¸Ðµ Ð½Ð° Ð²ÐµÑ€Ñ…Ð½Ð¸Ð¹ ÑƒÑ€Ð¾Ð²ÐµÐ½ÑŒ Ð¸Ð½Ð¶ÐµÐºÑ‚Ð¾Ñ€Ð°
            vNext = torch.tensor([0.0, 0.0])
        else:
            vNext = arrayEpoxyLevel[LevelMin+1] - arrayEpoxyLevel[LevelMin]
        if (LevelMin == 0): # ÐžÐ³Ñ€Ð°Ð½Ð¸Ñ‡ÐµÐ½Ð¸Ðµ Ð½Ð° Ð½Ð¸Ð¶Ð½Ð¸Ð¹ ÑƒÑ€Ð¾Ð²ÐµÐ½ÑŒ Ð¸Ð½Ð¶ÐµÐºÑ‚Ð¾Ñ€Ð°
            vPrev = torch.tensor([0.0, 0.0])
        else:
            vPrev = arrayEpoxyLevel[LevelMin] - arrayEpoxyLevel[LevelMin-1]

        cos_next = cos_sim(vSrc,vNext)
        cos_prev = cos_sim(vSrc,vPrev)

        #print ("Lvl: ", LevelMin+1, "\tXY next: ", arrayEpoxyLevel[LevelMin+1], "\tvNext: ", vNext, "\tL2: ", torch.linalg.vector_norm(vNext), "\tCos2Next: ", cos_next)
        #print ("Lvl: ", LevelMin, "\tXY el_c: ", arrayEpoxyLevel[LevelMin], "\tvSrc: ", vSrc, "\tL2: ", torch.linalg.vector_norm(vSrc))
        #print ("Lvl: ", LevelMin-1, "\tXY prev: ", arrayEpoxyLevel[LevelMin-1], "\tvPrev: ", vPrev, "\tL2: ", torch.linalg.vector_norm(vPrev), "\tCos2Prev: ", cos_prev)

        # Ð•ÑÐ»Ð¸ Ñ‚Ð¾Ñ‡ÐºÐ° Ð²Ñ‹ÑˆÐµ Ð±Ð»Ð¸Ð¶Ð°Ð¹ÑˆÐµÐ³Ð¾ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¾Ð²Ð°Ð½Ð½Ð¾Ð³Ð¾ ÑƒÑ€Ð¾Ð²Ð½Ñ, Ñ‚Ð¾ Ð´Ð¾Ð±Ð°Ð²Ð»ÑÐµÐ¼ Ð´Ð¾Ñ€Ð°ÑÑ‡Ð¸Ñ‚Ð°Ð½Ð½ÑƒÑŽ Ð´Ñ€Ð¾Ð±Ð½ÑƒÑŽ Ñ‡Ð°ÑÑ‚ÑŒ Ð¾Ð±ÑŠÐµÐ¼Ð°
        if cos_next > cos_prev:
            pr = pr_vec(vSrc, vNext)
            pr_ml = float (1 / torch.linalg.vector_norm(vNext) * pr)
        # Ð•ÑÐ»Ð¸ Ñ‚Ð¾Ñ‡ÐºÐ° Ð½Ð¸Ð¶Ðµ Ð±Ð»Ð¸Ð¶Ð°Ð¹ÑˆÐµÐ³Ð¾ ÐºÐ°Ð»Ð¸Ð±Ñ€Ð¾Ð²Ð°Ð½Ð½Ð¾Ð³Ð¾ ÑƒÑ€Ð¾Ð²Ð½Ñ, Ñ‚Ð¾ Ð²Ñ‹Ñ‡Ð¸Ñ‚Ð°ÐµÐ¼ Ð´Ð¾Ñ€Ð°ÑÑ‡Ð¸Ñ‚Ð°Ð½Ð½ÑƒÑŽ Ð´Ñ€Ð¾Ð±Ð½ÑƒÑŽ Ñ‡Ð°ÑÑ‚ÑŒ Ð¾Ð±ÑŠÐµÐ¼Ð°
        elif cos_next < cos_prev:
            pr = pr_vec(vSrc, vPrev)
            pr_ml = - float (1 / torch.linalg.vector_norm(vPrev) * pr)
        else:
            pr = 0
            pr_ml = 0
        #print ("Ð”Ð»Ð¸Ð½Ð° Ð¿Ñ€Ð¾ÐµÐºÑ†Ð¸Ð¸ Ð² Ð¿Ð¸ÐºÑÐµÐ»ÑÑ…: ", pr, "\tÐ”Ð»Ð¸Ð½Ð° Ð¿Ñ€Ð¾ÐµÐºÑ†Ð¸Ð¸ Ð² Ð¼Ð»: ", pr_ml)
        return LevelMin + pr_ml
    else:
        return None
```
